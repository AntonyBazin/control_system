```python
def calc_bias(lst):
  mean = np.mean(lst)
  print(f'Mean: {mean}, len: {len(lst)}')
  alpha = 0.95
  t_coef = 2.0086
  dx = t_coef * np.std(lst, ddof=1) / np.sqrt(len(lst))
  print(f'Result: ({np.mean(lst)}+-{dx:.2f}), alpha = {alpha}')
  return np.mean(lst), dx, alpha
```
